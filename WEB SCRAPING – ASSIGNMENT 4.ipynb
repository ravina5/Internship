# ANS.7

```python
driver = webdriver.Chrome()
driver.get("https://www.imdb.com/list/ls095964455/")
```

```python
ser_na = []

name= driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]')
for i in name:
    data = i.text
    ser_na.append(data)
    
```

```python
series_name_year = []
 
Genre_Run_time = []
Run_time = []
Ratings = []
Votes = []
for entry in ser_na:
    
    values = entry.split('\n')
    
    
    series_name_year.append(values[0])
    Genre_Run_time.append(values[1])
    Ratings.append(values[2])
    Votes.append(values[6])
   
```

```python
Genre = []
Run_time = []


for item in Genre_Run_time:
    
    components = [part.strip() for part in item.split("|")]
    
    genre = components[2]
    runtime = components[1]
    
    
    Genre.append(genre)
    Run_time.append(runtime)
```

```python
all_votes = [item.replace('Votes: ', '') for item in Votes ]
```

```python
series_names = []
years = []
for item in series_name_year:
    
    parts = item.split('(')
    
    name = parts[0].split('. ')[1]  
    year = parts[1].rstrip(')')  

    
    series_names.append(name)
    years.append(year)
```

```python
df7 = pd.DataFrame({'series_names':series_names,'years':years,'Genre':Genre,'Run_time':Run_time,'Ratings':Ratings,'all_votes':all_votes})
df7
```

```python
driver = webdriver.Chrome()
driver.get("https://archive.ics.uci.edu/")
```

```python
view_datasets =driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]")
view_datasets.click()
```

```python
row_page =driver.find_element(By.XPATH,'//select[@class="select-primary select select-sm rounded-full"]/option[5]')
row_page.click()
```

```python
datasets_url =[]
start = 0
end = 26
for page in range (start,end):
    url = driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
    for i in url:
        datasets_url.append(i.get_attribute('href'))
    next_button=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')    
    next_button.click()
    time.sleep(1)
```

```python
len(datasets_url)
```

```python
Dataset_name = []
Data_type = []
Task = []
Attribute_type = []
No_of_instances = []
No_of_attribute = []
Year = []

for i in datasets_url:
    driver.get(i)
    time.sleep(1)
    try:
        Name =driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/div/h1')
        Dataset_name.append(Name.text)
    except NoSuchElementException:
        Dataset_name.append('Not present')
        


    try:
        Data =driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[1]/p')
        Data_type.append(Data.text)
    except NoSuchElementException:
        Data_type.append('Not present')
        
        
        

    try:
        t =driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[3]/p')
        Task.append(t.text)
    except NoSuchElementException:
        Task.append('Not present')    
        
        
        


    try:
        a =driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[4]/p')
        Attribute_type.append(a.text)
    except NoSuchElementException:
        Attribute_type.append('Not present')     
        
        
        
    try:
        no =driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[5]/p')
        No_of_instances.append(no.text)
    except NoSuchElementException:
        No_of_instances.append('Not present')        
        
        
        
    try:
        no_of =driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[6]/p')
        No_of_attribute.append(no_of.text)
    except NoSuchElementException:
        No_of_attribute.append('Not present')  
        
      
    
    try:
        yy =driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/h2')
        Year.append(yy.text)
    except NoSuchElementException:
        Year.append('Not present')  
        
```

```python
df8 = pd.DataFrame({'Dataset_name':Dataset_name,'Data_type':Data_type,'Task':Task,'Attribute_type':Attribute_type,'No_of_instances':No_of_instances,'No_of_attribute':No_of_attribute,'Year':Year})
df8
```
