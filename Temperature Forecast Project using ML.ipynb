```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
```

```python
x = data1_copy.drop('Next_Tmax',axis = 1)
y = data1_copy['Next_Tmax']
```

```python
#Scaled the features
scalar = StandardScaler() 
x_scaled = scalar.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y, test_size= 0.25, random_state = 14)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

def metric_score(clf, x_train, x_test, y_train, y_test, train=True):
    if train:
        y_pred = clf.predict(x_train)
        print("\n============Train Result================")
        mae = mean_absolute_error(y_train, y_pred)
        mse = mean_squared_error(y_train, y_pred)
        r2 = r2_score(y_train, y_pred)
        print("Mean Absolute Error : ",mae)
        print("Mean Squared Error : ",mse)
        print("R-squared (R2) Score : ",r2)

    elif not train:
        y_pred = clf.predict(x_test)
        print("\n============Test Result================")
        mae = mean_absolute_error(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        print("Mean Absolute Error : ",mae)
        print("Mean Squared Error : ",mse)
        print("R-squared (R2) Score : ",r2)
```

### KNeighborsRegressor

```python
knn= KNeighborsRegressor()
knn.fit(x_train, y_train)
```

```python
metric_score(knn,x_train,x_test,y_train,y_test,train=True)        
metric_score(knn,x_train,x_test,y_train, y_test,train=False)        
```

### LinearRegression

```python
lin= LinearRegression() 
lin.fit(x_train, y_train)
```

```python
metric_score(lin,x_train,x_test,y_train,y_test,train=True)        
metric_score(lin,x_train,x_test,y_train, y_test,train=False)        
```

### DecisionTreeRegressor

```python
dtr= DecisionTreeRegressor()
dtr.fit(x_train, y_train)
```

```python
metric_score(dtr,x_train,x_test,y_train,y_test,train=True)        
metric_score(dtr,x_train,x_test,y_train, y_test,train=False)        
```

### AdaBoostRegressor

```python
ada = AdaBoostRegressor()
ada.fit(x_train,y_train)
```

```python
metric_score(ada,x_train,x_test,y_train,y_test,train=True)        
metric_score(ada,x_train,x_test,y_train, y_test,train=False)        
```

#### knn model has the best score so I will select this model.
